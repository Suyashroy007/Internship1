```python
import requests
from bs4 import BeautifulSoup
  
URL = "https://en.wikipedia.org/wiki/Main_Page"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify())
```

```python
heading = soup.find_all(['h1','h2','h3','h4','h5','h6'])
heading
```
